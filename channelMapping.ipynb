```python
import csv
import numpy as np
import os
import utils
```

```python
def read_train_data(file_name):
    with open(file_name, 'r', newline='') as f:
        lines = csv.reader(f)
        data = []
        for line in lines:
            data.append(line)

    data = np.array(data).astype(np.float64)
    return data


def save_data(data, filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)

def reconstruction(input_path, output_path, model, srate):
    for f in os.listdir(input_path):
        print(input_path+f)
        old_data = read_train_data(input_path+f)
        new_data = np.zeros((30, old_data.shape[1]))
        #old_idx = [ 1, 1, 1, 4, 1, 6, 1, 1, 1, 1, 1, 1, 13, 14, 1, 16, 17, 1, 1, 1, 1, 1, 1, 24, 1, 26, 1, 1, 1, 1]
        old_idx = [4, 6, 13, 14, 16, 17, 24, 26]
        # channel mapping
        for j in range(8):
            new_data[old_idx[j]-1, :] = old_data[j, :]

        #output_name = f[:-4] + "_" + str(g_time) + f[-4:]
        # step2: Signal reconstruction
        for m in model:
            preprocess_data = utils.preprocessing(new_data, srate)
            print(f[:-4] + "_" + str(m))
            output_name = f[:-4] + "_" + str(m)
            # step2: Signal reconstruction
            utils.reconstruct(m, preprocess_data, output_path+output_name)
def slicedata(input_path, output_path, srate):
    for f in os.listdir(input_path):
        print(input_path+f)
        old_data = read_train_data(input_path+f)
        new_data = np.zeros((30, old_data.shape[1]))
        #old_idx = [ 1, 1, 1, 4, 1, 6, 1, 1, 1, 1, 1, 1, 13, 14, 1, 16, 17, 1, 1, 1, 1, 1, 1, 24, 1, 26, 1, 1, 1, 1]
        old_idx = [4, 6, 13, 14, 16, 17, 24, 26]
        # channel mapping
        for j in range(8):
            new_data[old_idx[j]-1, :] = old_data[j, :]

        #output_name = f[:-4] + "_" + str(g_time) + f[-4:]
        # step2: Signal reconstruction
        preprocess_data = utils.preprocessing(new_data, srate)
        output_name = f[:-4]
        # step2: Signal reconstruction
        utils.sclicedata(preprocess_data, output_path+output_name)

def combineData(EC, CPT, output_path):
    for f in os.listdir(EC):
        try:
            ECdata = np.load(EC+f)
            CPTdata = np.load(CPT+f)
            combine = np.concatenate((ECdata, CPTdata), axis=2)
            np.save(output_path+f[:-4], combine)
        except:
            print(f'{f} is not in CPT')
        
```
